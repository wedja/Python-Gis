```python
# abrir o rquivo municipio
dados = gpd.read_file('./arquivos_vetoriais/municipios.shp')
dados
```

```python
dados.head()
```

```python
dados[:5].plot()
```

```python
#Realizar um filtro e salvar o arquivo em novo arquivo chamado mucipios_mt
filtro = dados['uf'] == 'MT'
municipio_mt = dados.loc[filtro]
municipio_mt.to_file('./arquivos_vetoriais/municipio_mt.shp')
```

```python
area_mt = gpd.read_file('./arquivos_vetoriais/municipio_mt.shp')
area_mt
```

```python
#Calcular a area em graus
municipio_mt['area'] = municipio_mt['geometry'].area
municipio_mt
```

```python
import numpy as np
municipio_mt['classe'] = np.where(municipio_mt['area'] < 0.1,1,0)
municipio_mt
```

```python
municipio_mt
```

```python
municipio_mt.crs #tipo de dados
```

```python
#setando a cooredenada geografica
municipio_mt_crs = municipio_mt.set_crs(epsg=4674, allow_override=True)
```

```python
municipio_mt_crs
```

```python
municipio_mt_crs.crs
```

```python
#Conersão de coordenadas para metros
municipio_mt['geometry'] = municipio_mt['geometry'].to_crs(epsg=31981)
municipio_mt
```

```python
import geopandas as gpd
import os
import matplotlib.pyplot as plt
```

```python
mun = gpd.read_file('./arquivos_vetoriais/municipios.shp')
mun
```

```python
aeroportos = gpd.read_file('./arquivos_vetoriais/aeroportos.shp')
aeroportos
```

```python
mun_pr = mun[mun['uf'] == 'PR']
aero_pr = gpd.sjoin(aeroportos, mun_pr, predicate='within')
aero_pr.plot()
```
